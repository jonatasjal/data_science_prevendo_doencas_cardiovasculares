#### 1.3.4 Variável oldpeak

**Depressão de ST induzida por exercício em relação ao repouso**

Descreve uma condição em que há uma alteração específica no ECG, representada pela depressão do segmento ST, que ocorre como resultado de atividade física (exercício), em comparação com o estado de repouso. Este tipo de análise é muitas vezes utilizado para avaliar a saúde cardiovascular e identificar possíveis problemas cardíacos.

- Depressão de ST: Refere-se a uma alteração no segmento ST de um eletrocardiograma (ECG). O segmento ST é a parte do ECG que representa o intervalo entre o final da contração ventricular (complexo QRS) e o início da repolarização ventricular. A depressão de ST pode ser um sinal de comprometimento cardíaco.

- Induzida por exercício: Sugere que essa depressão de ST ocorre como resultado de atividade física ou exercício. Muitas vezes, durante um teste de esforço ou um exame de estresse, os pacientes são submetidos a atividades físicas para avaliar a resposta do coração ao esforço.

- Em relação ao repouso: Indica que a comparação está sendo feita entre os resultados durante o exercício e os resultados em um estado de repouso. Ou seja, a depressão de ST que está sendo observada ocorre durante ou após o exercício, em comparação com a linha de base do ECG em repouso.

```python
data['oldpeak'].unique()
```

> Não encontramos uma documentação sobre a escala do dataset, e essa variável possui uma correlação média negativa em torno de -43%. Acreditamos que um valor como 1. indica uma depressão de 1 mm no segmento ST durante ou após o exercício em comparação com o repouso; 3.1 indica uma depressão de 3.1 mm, e assim por diante. 

> A magnitude da depressão do segmento ST pode ser um indicativo de estresse no músculo cardíaco durante o exercício.

#### 1.3.9 Variável serum_cholestoral

O colesterol sérico em mg/dL representa a quantidade de colesterol no sangue, medido em miligramas por decilitro. Esse valor é crucial para avaliar a saúde cardiovascular, pois o colesterol desempenha papéis essenciais no corpo. Níveis elevados podem indicar riscos para doenças cardíacas. 

Existem três principais tipos de colesterol avaliados no sangue:

Colesterol Total: Representa a soma de todos os tipos de colesterol no sangue, incluindo o colesterol de lipoproteínas de baixa densidade (LDL) e o colesterol de lipoproteínas de alta densidade (HDL).

LDL (Low-Density Lipoprotein - Lipoproteína de Baixa Densidade): Comumente conhecido como "colesterol ruim", níveis elevados de LDL estão associados ao acúmulo de placas nas artérias, aumentando o risco de doenças cardiovasculares.

HDL (High-Density Lipoprotein - Lipoproteína de Alta Densidade): Conhecido como "colesterol bom", o HDL ajuda a remover o excesso de colesterol do sangue, reduzindo o risco de doenças cardíacas.



```python
plt.figure(figsize=(12, 6))
sns.boxplot(x=data['serum_cholestoral'], orient='h')
plt.title('Boxplot de Colesterol Sérico mg/dl')
plt.show()
```

Essa variável tem algumas peculiaridades:

### 3.1 RandomForest

O Random Forest é um algoritmo do pacote **ensemble learning** e nós usamos esse primeiramente, pois engloba a construção de um modelo forte a partir da combinação de vários modelos mais fracos. No caso do Random Forest, os modelos mais fracos são árvores de decisão.

Para quem não conhece, algumas das principais caracteríssticas do Random Forest são:

1. **Árvores de Decisão Base:** o algoritmo constrói várias árvores de decisão independentes durante o treinamento e cada árvore é treinada em uma amostra aleatória do conjunto de dados, e essa amostra é geralmente obtida por meio de bootstrapping (amostragem com substituição).

2. **Amostragem Aleatória de Recursos:** além de amostrar aleatoriamente observações, ele também faz uma seleção aleatória de um subconjunto de recursos em cada divisão de árvore, o que ajuda a tornar as árvores mais independentes e diversificadas.

3. **Votação para Previsões:** durante a fase de previsão, as previsões de cada árvore são consideradas atraves de uma "votação" que é realizada para determinar a previsão final. <br>
**Importante:** Para problemas de *classificação*, a classe mais frequente é escolhida, enquanto para problemas de *regressão*, a média das previsões é tomada.

4. **Robustez e Generalização:** ele é mais robusto em relação a overfitting devido à combinação de várias árvores. Ou seja,  ele geralmente produz modelos mais estáveis e capazes de generalizar bem para novos dados.

5. **Importância de Recursos:** o algoritmo fornece uma medida da importância de cada recurso no processo de tomada de decisão. Essa informação pode ser valiosa para entender quais características são mais relevantes para o problema em questão.

6. **Versatilidade:** por fim, ele pode ser aplicado a uma variedade de tarefas, incluindo classificação e regressão. Também é menos sensível a hiperparâmetros em comparação com uma única árvore de decisão.

Em resumo, o Random Forest é muito utilizado devido à sua eficácia, versatilidade e facilidade de uso.

```python
# Criando uma instância
rf_classifier = RandomForestClassifier()
rf_classifier.fit(x_train, y_train)
```

- Precision: proporção de verdadeiros positivos em relação a todos os positivos previstos
- Recall (ou Sensibilidade): é a proporção de verdadeiros positivos em relação a todos os verdadeiros positivos possíveis

#### 3.1.2 Curva AUC e Curva ROC

Aqui queremos descobrir o desempenho do modelo em termos de sua capacidade de distinguir entre classes positivas e negativas.

- A Curva AUC (Área Sob a Curva) é uma métrica utilizada na avaliação de modelos de classificação para mensurar esse desempemnho, onde a área sob a curva representa a probabilidade de que o modelo classifique uma instância positiva aleatória mais alta do que uma instância negativa aleatória.

- A Curva ROC é um gráfico que mostra como o desempenho de um modelo de classificação varia conforme ajustamos os critérios de decisão. É útil em problemas de classificação binária para avaliar a capacidade do modelo em diferentes níveis de sensibilidade e especificidade. A curva é criada plotando a taxa de verdadeiros positivos (sensibilidade) em relação à taxa de falsos positivos (1 - especificidade) para diferentes valores de limiar de decisão.

```python
from sklearn.metrics import roc_curve, auc

# Curva ROC e AUC
y_scores = rf_classifier.predict_proba(x_test)[:, 1]
fpr, tpr, thresholds = roc_curve(y_test, y_scores)
roc_auc = auc(fpr, tpr)

# Plote a Curva ROC
plt.figure(figsize=(12, 8))
plt.plot(fpr, tpr, color='lightgreen', lw=2, label=f'Curva ROC (AUC = {roc_auc:.2f})')
plt.plot([0, 1], [0, 1], color='green', lw=2, linestyle='--')
plt.xlabel('Taxa de Falsos Positivos')
plt.ylabel('Taxa de Verdadeiros Positivos')
plt.title('Curva ROC')
plt.legend(loc='lower right')
plt.show()
```

Quanto mais a linha azul se aproxima do canto superior esquerdo, melhor o modelo é em distinguir entre as classes. Por outro lado, se a linha estiver mais próxima da diagonal no meio do gráfico (subindo da parte inferior esquerda até a superior direita), o modelo tem menos habilidade em diferenciar entre as classes.

### 3.2 AdaBoost

Já Adaboost, ou "Adaptive Boosting", é como ter um grupo de especialistas em diferentes assuntos trabalhando juntos para tomar uma decisão e seu funcionamento leva em conta alguns conceitos:

1. **Início Fraco (Weak Starters):** Imagine que você está tentando aprender várias matérias na escola, e cada matéria tem um professor diferente. Alguns professores são melhores em certos tópicos do que outros. No início, você começa com um professor (um modelo fraco) para cada matéria.

2. **Foco nas Dificuldades (Focus on Mistakes):** Agora, você faz um teste, e onde você comete erros, esses erros são identificados. Aqui, o Adaboost foca nos erros cometidos pelo modelo fraco.

3. **Maior Atenção aos Erros (Boosting Mistakes):** O Adaboost dá mais atenção aos erros cometidos, então, na próxima vez, você obtém a ajuda de todos os professores, mas com mais ênfase nas matérias em que você teve mais dificuldade (modelos que cometeram mais erros).

4. **Decisões em Grupo (Group Decision):** Agora, todos os professores votam juntos sobre a resposta correta para cada pergunta no teste. Aqueles que são bons nas matérias em que você teve mais dificuldade têm mais influência nas decisões.

5. **Ciclos de Refinamento (Repeating the Process):** O processo se repete, dando mais atenção aos erros anteriores a cada rodada. Cada vez, o Adaboost ajusta o foco, aprendendo e melhorando com base nos erros cometidos.

6. **Modelo Forte (Strong Model):** Após várias rodadas, você tem um grupo de professores que, juntos, formam um modelo forte. Cada um contribui mais nas áreas em que é especialista, e todos trabalham em conjunto para responder melhor.

Em resumo, o Adaboost é como ter vários especialistas corrigindo seus erros, ajustando-se ao que você tem mais dificuldade, até que todos trabalhem juntos para criar um modelo mais forte e preciso.

### 3.3 Gradient Boosting

Gradient Boosting é um algoritmo que também utiliza a ideia de construir um modelo forte a partir de modelos mais fracos, só que ele o faz de uma maneira um pouco diferente. Eu vou usar uma analogia que me fez compreender de maneira mais rápida, que é a analogia do subir de uma montanha.

1. **Learning from Mistakes:** ao subirmos a montanha, a cada passo, nós avaliamos o quanto estamos longe do topo. De início, não sabemos a melhor direção para subir, mas aprendemos com cada passo, ajustando a trajetória com base nos erros anteriores.

2. **Modelo Inicial:** no começo, fazemos uma estimativa inicial de como subir a montanha. Essa é a nossa primeira tentativa (um modelo fraco), que pode não ser muito precisa.

3. **Identificação dos Erros:** verificamos onde erramos na primeira tentativa, ou seja, onde você estimamos incorretamente a distância até o topo, que são os lugares que precisamos melhorar.

4. **Modelos Adicionais:** agora, você construimos modelos adicionais, com o objetivo de corrigir os erros identificados anteriormente. Cada novo modelo se concentra nas áreas onde o modelo anterior teve mais dificuldade.

5. **Soma Ponderada:** combinamos os modelos e os somamos, mas com um peso maior para aqueles que tiveram mais sucesso em corrigir os erros para criar uma estimativa mais precisa.

6. **Ciclos de Refinamento:** repetimos o processo, iterativamente. Ou seja, a cada ciclo, novos modelos são construídos para corrigir as deficiências do conjunto atual.

7. **Convergência para um Modelo Forte:** com o tempo, ajustamos os modelos fracos e os combinados de forma a convergir para um modelo forte que é muito preciso na predição.
